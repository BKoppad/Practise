Draw a plot with x and y axis

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(0,10,1)
x
```

```python
y=np.arange(0,10,1)
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('X vs Y')
```
